```python
data['liked'] = data['liked'].astype(bool)
data['progress'] = data['progress'].astype(float)
data['course_rating'] = data['course_rating'].astype(float)
data['student_score'] = data['student_score'].astype(float)
data['share_count'] = data['share_count'].astype(int)
```

```python
data['review_text'] = data['review_text'].fillna('')
```

```python
# We choose 'course_rating' as the metric for CF
user_course_matrix = data.pivot_table(
    index='user_id',
    columns='course_id',
    values='course_rating',
    aggfunc='mean'  # if multiple enrollments for same user-course, average them
).fillna(0)

# Convert to NumPy for similarity
user_matrix = user_course_matrix.values
```
